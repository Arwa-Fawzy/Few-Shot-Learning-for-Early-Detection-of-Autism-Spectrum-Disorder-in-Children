```python
class PrototypicalNetworks(nn.Module):
    def __init__(self, backbone: nn.Module):
        super(PrototypicalNetworks, self).__init__()
        self.backbone = backbone

    def forward(
        self,
        support_images: torch.Tensor,
        support_labels: torch.Tensor,
        query_images: torch.Tensor,
    ) -> torch.Tensor:
        
        # Extract features
        z_support = self.backbone.features(support_images)
        z_query = self.backbone.features(query_images)
        
        # Apply the first two layers of the classifier
        z_support = self.backbone.classifier[:2](z_support)
        z_query = self.backbone.classifier[:2](z_query)

        # Flatten the features
        z_support = z_support.view(z_support.size(0), -1)
        z_query = z_query.view(z_query.size(0), -1)

        # Apply the next two layers of the classifier
        z_support = self.backbone.classifier[2:4](z_support)
        z_query = self.backbone.classifier[2:4](z_query)
        
        # Normalize the embeddings
        z_support = F.normalize(z_support, p=2, dim=-1)
        z_query = F.normalize(z_query, p=2, dim=-1)

        # Number of unique classes in the support set
        n_way = len(torch.unique(support_labels))
        
        # Compute the prototypes
        z_proto = torch.stack(
            [
                z_support[torch.nonzero(support_labels == label, as_tuple=True)].mean(0)
                for label in torch.unique(support_labels)
            ]
        ).squeeze(1)
        
        # Compute cosine similarity
        dists = F.linear(z_query, z_proto)
        
        # Transform distances into classification scores
        scores = dists
        return scores

# Load the MobileNetV2 model with pretrained weights
convolutional_network = mobilenet_v2(pretrained=True).cuda()

# Define L2 regularization
l2_lambda = 0.015
def l2_reg(m):
    if isinstance(m, nn.Linear):
        nn.init.xavier_uniform_(m.weight)
        m.weight = nn.Parameter(m.weight - l2_lambda * m.weight)
        if m.bias is not None:
            nn.init.constant_(m.bias, 0.0)

# Modify the classifier of the MobileNetV2 model
convolutional_network.classifier = nn.Sequential(
    nn.AdaptiveAvgPool2d(1),  # Reduce to (batch_size, 1280, 1, 1)
    nn.Flatten(),  # Flatten to (batch_size, 1280)
    nn.Dropout(0.4),
    nn.Linear(convolutional_network.last_channel, 128),  # Adjusted to match the flattened output
    nn.ReLU(),
    nn.Linear(128, 64),  # Intermediate feature layer
    nn.ReLU(),
    nn.Linear(64, 2),  # Assuming 2 is the number of classes
    nn.Softmax(dim=1)
)

# Apply L2 regularization to the linear layer
convolutional_network.classifier[3].apply(l2_reg)

# Create the Prototypical Networks model
model = PrototypicalNetworks(convolutional_network).cuda()
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(model.parameters(), lr=0.0001, momentum=0.9)


def val_fit(
    support_images: torch.Tensor,
    support_labels: torch.Tensor,
    query_images: torch.Tensor,
    query_labels: torch.Tensor,
) -> float:
    optimizer.zero_grad()
    classification_scores = model(
        support_images.cuda(), support_labels.cuda(), query_images.cuda()
    )

    loss = criterion(classification_scores, query_labels.cuda())
    optimizer.step()

    return loss.item()
```

```python
def evaluate_on_one_task(
    support_images: torch.Tensor,
    support_labels: torch.Tensor,
    query_images: torch.Tensor,
    query_labels: torch.Tensor,
) -> [int, int]:
    """
    Returns the number of correct predictions of query labels, and the total number of predictions.
    """
    return (
        torch.max(
            model(support_images.cuda(), support_labels.cuda(), query_images.cuda())
            .detach()
            .data,
            1,
        )[1]
        == query_labels.cuda()
    ).sum().item(), len(query_labels)
```

```python
def evaluate(data_loader: DataLoader):
    # eval mode affects the behaviour of some layers (such as batch normalization or dropout)
    # no_grad() tells torch not to keep in memory the whole computational graph (it's more lightweight this way)
    model.eval()
    with torch.no_grad():
        for episode_index, (
            support_images,
            support_labels,
            query_images,
            query_labels,
            class_ids,
        ) in tqdm(enumerate(data_loader), total=len(data_loader)):
            
            loss_value = val_fit(support_images, support_labels, query_images, query_labels)
            validation_losses.append(loss_value)

            correct, total = evaluate_on_one_task(
                support_images, support_labels, query_images, query_labels
            )
            
            validation_accuracies.append((100 * correct/total))
            

    print(
        f"Model tested on {len(data_loader)} tasks. Accuracy: {(100 * correct/total):.2f}%"
    )
```

```python
N_TRAINING_EPISODES = 40000
N_VALIDATION_TASKS = 500

# Training set
train_set.get_labels = lambda: [instance[1] for instance in train_set.samples]
train_sampler = TaskSampler(
    train_set, n_way=N_WAY, n_shot=N_SHOT, n_query=N_QUERY, n_tasks=N_TRAINING_EPISODES)
train_loader = DataLoader(
    train_set,
    batch_sampler=train_sampler,
    num_workers=2,
    pin_memory=True,
    collate_fn=train_sampler.episodic_collate_fn,
)

# Validation set
valid_set.get_labels = lambda: [instance[1] for instance in valid_set.samples]
validation_sampler = TaskSampler(
    valid_set, n_way=N_WAY, n_shot=N_SHOT, n_query=N_QUERY, n_tasks=N_VALIDATION_TASKS
)
validation_loader = DataLoader(
    valid_set,
    batch_sampler=validation_sampler,
    num_workers=2,
    pin_memory=True,
    collate_fn=validation_sampler.episodic_collate_fn,
)
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(model.parameters(), lr=0.0001, momentum=0.9)


def fit(
    support_images: torch.Tensor,
    support_labels: torch.Tensor,
    query_images: torch.Tensor,
    query_labels: torch.Tensor,
) -> float:
    optimizer.zero_grad()
    classification_scores = model(
        support_images.cuda(), support_labels.cuda(), query_images.cuda()
    )

    loss = criterion(classification_scores, query_labels.cuda())
    loss.backward()
    optimizer.step()

    return loss.item()
```

```python
log_update_frequency = 10

model.train()
with tqdm(enumerate(train_loader), total=len(train_loader)) as tqdm_train:
    for episode_index, (
        support_images,
        support_labels,
        query_images,
        query_labels,
        _,
    ) in tqdm_train:
        loss_value = fit(support_images, support_labels, query_images, query_labels)
        training_losses.append(loss_value)
        correct, total = evaluate_on_one_task(
                support_images, support_labels, query_images, query_labels
            )
        training_accuracies.append((100 * correct/total))
        if episode_index % log_update_frequency == 0:
            tqdm_train.set_postfix(loss=sliding_average(training_losses, log_update_frequency))
```

```python
def plot_confusion_matrix(y_true, y_pred, labels, fontsize=20):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(8, 6))  
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=labels, yticklabels=labels, annot_kws={"size": fontsize})
    plt.title('Confusion Matrix', fontsize=fontsize-4)
    plt.xlabel('Predicted', fontsize=fontsize-4)
    plt.ylabel('True', fontsize=fontsize-4)
    plt.xticks(fontsize=fontsize)
    plt.yticks(fontsize=fontsize)
    plt.show()
```
